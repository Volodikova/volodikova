# Предсказываем качество вина



```python
#Загузка библиотек
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

## Загружаем данные

Собраны данные:
- fixed acidity
- volatile acidity
- citric acid
- residual sugar
- chlorides
- freesulfur dioxide
- total sulfur dioxide
- density
- pH
- sulphates
- alcohol
- quality

```python
# Загружаем данные из файла winequality-white.csv в переменную df
```

```python
df=pd.read_csv('winequality-white.csv',sep=';')
```

```python
df.head()
```

## Смотрим статистику, что нет пропусков и отсуствующих значений

```python
df.info()
```

```python
df.describe()
```

```python
# повезло с датасетом, все данные строковые и не пустые
```

```python
df.head()
```

## Формируем признаки и целевую переменную

```python
## формируем признаки и целнвую переменную
```

```python
target_value=df['quality']
```

```python
x=df[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']]
y=target_value
```

```python
x
```

```python
y
```

```python
## разделяем данные на обучающую и тестовую выборки
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
x_train
```

```python
y_train
```

```python
#обучение модели
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)
lr.coef_
```

## Получаем предсказание и оцениваем качество

```python
# Получить предсказание
```

```python
lr.predict(x_test)
```

```python
#оценка на глаз
```

```python
pd.DataFrame([y_test.values,lr.predict(x_test)]).T
```

```python
# Оценка качества при помощи метода MSE для тестовой и обучающей выборки
```

```python
mean_squared_error(y_test,lr.predict(x_test))
```

```python
mean_squared_error(y_train,lr.predict(x_train))
```
